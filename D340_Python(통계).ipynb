# 🔗 짝지은 두 집단의 비교 (Paired Sample t-test)

---

## 🎯 1. 개념

- **혼란변수(confounding variable)**는  
  원인변수와 결과변수 모두에 영향을 주어, 실제로는 인과관계가 없는데도  
  두 변수 사이에 상관이 있는 것처럼 보이게 만든다.

→ 따라서 혼란변수의 영향을 **통제(control)** 할 필요가 있다.  
하지만 실제 연구에서는 모든 혼란변수를 통제하기 어려우므로,  
**동일한 집단 내 투약 전·후를 비교하는 방법(대응표본 설계)**을 활용할 수 있다.

---

## 🧠 2. 동일 집단 내 전·후 비교의 장점

- **대상자 수 절감:**  
  두 집단을 따로 모집할 필요가 없으므로, 필요한 대상자 수를 절반으로 줄일 수 있다.

- **혼란변수 통제 가능:**  
  동일한 대상이므로, 개인차(성별, 나이, 체질 등)에 의한 혼란효과를 자동으로 통제할 수 있다.

---

## ⚖️ 3. 짝지은 두 표본 검정의 원리

짝지은 두 표본 검정(Paired t-test)은  
각 대상자별로 **“전(before) 값 - 후(after) 값”**의 차이를 계산하고,  
이 차이값으로 만든 **새로운 분포**를 이용해 검정을 수행한다.

- H₀ (귀무가설): 투여 전후 값의 평균 차이가 없다 (μ_d = 0)  
- H₁ (대립가설): 투여 전후 값의 평균 차이가 있다 (μ_d ≠ 0)

여기서 `d`는 (전 - 후)의 차이값을 의미한다.

---

## 🔍 4. 검정 절차

1. 각 대상자별로 **(투여 전 - 투여 후)** 값을 계산한다.  
2. 이 차이값들의 **평균(mean)** 과 **표준편차(SD)** 를 구한다.  
3. 차이값이 **정규분포를 따른다고 가정**하고, t-검정을 수행한다.  
   - 표준화한 t값이 95% 신뢰구간 밖에 있으면 → 귀무가설 기각  
   - 즉, **투여 전후에 유의미한 차이가 있다**고 판단한다.

---

## 📈 5. 직접 추정 방식 (신뢰구간 이용)

예를 들어, 투여 전후 차이값의 95% 신뢰구간이 다음과 같다고 하자.

> 2.1 ≤ μ_d ≤ 9.9

→ 신뢰구간에 **0을 포함하지 않으므로**,  
귀무가설(μ_d = 0)을 **기각**한다.  
즉, **투여 전후 평균에 통계적으로 유의한 차이가 있다**고 결론 내린다.

---

## ✅ 요약

| 구분 | 특징 |
|------|------|
| 검정 이름 | 짝지은 두 표본 t-검정 (Paired t-test) |
| 사용 상황 | 동일한 대상의 전후 비교, 혹은 짝이 맞는 두 데이터 |
| 귀무가설 | 전후 평균 차이가 없다 (μ_d = 0) |
| 대립가설 | 전후 평균 차이가 있다 (μ_d ≠ 0) |
| 장점 | 개인차 및 혼란변수 통제 가능, 표본 수 절감 |
| 판정 기준 | p-value < 0.05 또는 신뢰구간에 0 미포함 시 → 유의한 차이 |

- 모수검정과 비교했을 때는 비모수검정은 통계적 유의성이 나오기가 더 힘들다. 이에 비모수보다 모수 검정을 더 하고싶은데, 정규분포를 가정할 수 있는 상황에만 모수 검정을 해야한다.

- 귀무가설: 이전에 발생하던 경향에서 변함이 없다, 우연에 의해 언제든지 일어날 수 있는 일이다, 이 사건이 일어나는 것은 매우 당연하다, 이 사건이 이어날 확률은 낮지 않다.

- 대립가설: 이전에 발생하던 경향에서 변함이 일어났다, 우연에 의해 일어날 수 없는 일이다, 이 사건이 일어나는 것은 당연하지 않다, 이 사건이 일어날 확률은 극히 낮다
