# 🧮 통계검정을 위한 귀무가설 설정 및 절차 정리

---

## 🎯 1. 귀무가설(H₀)과 대립가설(H₁) 설정

- **양측 검정 (Two-tailed test)**  
  - H₀: 두 집단의 평균이 같다 (μ₁ = μ₂)  
  - H₁: 두 집단의 평균이 다르다 (μ₁ ≠ μ₂)  
  → 두 집단 간 **평균 차이가 존재하는지 여부**를 검정할 때 사용

- **좌측 검정 (Left-tailed test)**  
  - H₀: 집단1의 평균이 더 작지 않다 (μ₁ ≥ μ₂)  
  - H₁: 집단1의 평균이 더 작다 (μ₁ < μ₂)  
  → 집단1의 평균이 **더 작은지** 확인할 때 사용

- **우측 검정 (Right-tailed test)**  
  - H₀: 집단1의 평균이 더 크지 않다 (μ₁ ≤ μ₂)  
  - H₁: 집단1의 평균이 더 크다 (μ₁ > μ₂)  
  → 집단1의 평균이 **더 큰지** 확인할 때 사용

📌 즉, **두 집단의 평균 차이 여부**를 검정할 때는 **양측 검정**을 사용한다.

## 📊 2. 정규성 여부에 따른 검정법 선택

| 표본 크기(n) | 분포 형태 | 적용 검정 |
|---------------|------------|-------------|
| n > 30 | 대체로 정규분포로 간주 가능 | ✅ **모수 검정 (t-test)** |
| n < 10 | 정규성 가정 어려움 | ❌ **비모수 검정 (Mann-Whitney U 등)** |
| 10 ≤ n ≤ 30 | 분포 형태 직접 확인 필요 | ⬇️ 아래 절차 참고 |

➡️ **표본 크기 10 ≤ n ≤ 30이거나 분포가 비대칭일 경우:**  
1. **히스토그램**으로 분포의 대칭성 확인 (왼쪽/오른쪽 치우침 여부)  
2. **정규성 검정(Shapiro-Wilk Test)** 수행  
   - p-value > 0.05 → 정규분포 가정 가능 → **모수 검정 (t-test)**  
   - p-value ≤ 0.05 → 정규분포 가정 불가 → **비모수 검정 (Mann-Whitney 등)**

## 🔍 3. 신뢰구간과 기각역 개념

- 신뢰수준 95%라면,  
  표본으로부터 계산된 평균이 **95% 신뢰구간(Confidence Interval)** 안에 있을 확률이 95%라는 뜻.  

- 표본평균이 **신뢰구간 밖(=유의수준 5% 영역)** 에 위치한다면,  
  귀무가설을 **기각**한다.

- 이 신뢰구간 바깥쪽, 즉 극단 영역을 **기각역(Rejection region)** 이라고 한다.

## ⚖️ 4. 두 확률변수의 평균 관계

두 확률변수 X, Y가 서로 **독립**일 때  
- E(X+Y) = E(X) + E(Y)  
즉, 각 분포의 평균을 더한 값이 합의 분포 평균과 같다.

## 💡 5. 두 집단 평균 차이 검정의 핵심

두 군 간 평균 차이를 검정하기 위해서는  
- **치료군의 평균(μ₁)**, **표준편차(s₁)**  
- **대조군의 평균(μ₂)**, **표준편차(s₂)**  
를 이용하여 다음 귀무가설을 세운다.

> H₀: μ₁ - μ₂ = 0  
> (두 집단의 평균 차이가 없다)

→ p-value가 유의수준(보통 0.05)보다 작으면  
귀무가설을 기각하고 **두 집단 간 평균 차이가 통계적으로 유의하다**고 결론 내린다.
표준화값이 95% 신뢰구간 안에 있으면 귀무가설을 기각하지 못한다.

μ₁-μ₂ = 0 이다 귀무가설로 가설검정을 할 수 있지만, μ₁-μ₂ 의 값을 직접 추정하는 방식도 가능하다.
