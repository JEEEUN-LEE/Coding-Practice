Q. 자기소개

안녕하세요, 보건정보통계학 석사 과정 지원자 이지은입니다.
저는 현재 국립암센터에서 항암제 위주 임상시험 데이터를 관리하고 있습니다. 
간호학을 전공하고, 연세대학교 산학협력단에서 응급의료시스템 개발 과제에 참여하며 의료데이터를 정제하고 분석한 경험이 있습니다. SAS 프로그램을 사용하여 국내 응급의료기관사용 현황에 대해 분석할 수 있었고, 이 경험을 통해
통계분석방법론에 대해 체계적으로 배우고자 하는 목표가 생겼습니다. 
이에 Python과 SAS SQL 언어를 꾸준히 익히고, 관련된 자격증들을 취득하였고, 또 온오프라인 교육을 통해서 데이터분석을 실습하고 현재 실무에도 적용하고 있습니다.
또한 올해 한국보건산업진흥원에서 주관한 바이오헬스 정책 아이디어 공모전에서 AI 기반 임상시험 대상자 모집 시스템을 주제로 장려상을 수상하기도 하였습니다.
저는 이번이 세번째 지원인데요, 데이터에 대한 흥미와, 그 의미를 밝혀내는 통계에 대한 열정이 있기에 포기하지 않고 도전할 수 있었습니다.
올해에는 꼭 입학하여서 통계분석까지 수행할 수 있는 보건의료전문가로 성장하고싶습니다.



Q. 불합격 이유, 불합격 이후 노력 어필? 

이전에는 현업 경험이 길지 않고, 하고 있는 업무와 석사과정을 연결할 수 있는 전문성이 조금 부족했던 것 같습니다.
그래서 결과발표 날로부터 매일 알고리즘 문제 풀이나 통계이론을 정리해 깃허브에 업로하며 꾸준히 기본기를 채워갔습니다. 또한 실무 적용을 높이기 위해서, 스터디 팀원들과 통계/머신러닝 책을 공부하며 직접 공부한 것으로 설명하는 방식으로 참여했고,
마지막으로는 체력관리가 무엇보다 중요하다 생각하여 매달 20km 이상씩 달리면서 체력을 늘렸습니다. 올해에는 두번의 10km 마라톤도 예정되어 있습니다.

이러한 과정에 통해 꾸준히 성장하는 습관을 유지했고, 그동안의 노력을 바탕으로 데이터 관리 뿐 아니라 통계분석과 연구 설계까지 이해하는 전문가로 한단계 더 성장하고 싶습니다.

- 1. 결과 발표 날 이후부터 매일매일 알고리즘 문제 풀이 또는 통계 이론에 대한 정리를 하여 깃허브에 업로드
  2. 실무에 적용하기 위해 적극적으로 스터디원과 통계/머신러닝 책을 공부하고 설명하는 참여
  3. 체력 강화를 위해 매달 20km 이상씩 달리며 올해에도 두번의 10km 마라톤

Q. 공모전 내용

말씀드린 바와 같이 AI기반 임상시험 대상자 모집 시스템을 주제로 하였습니다. 우선 병원끼리의 의료 데이터 통합 시스템 구축하고, 인공지능을 통해 선정/제외기준에 부합하는 대상자를 자동으로 찾아주는 플랫폼을 구축하는 것입니다.
제가 근무하는 곳에서는 연구자주도 임상시험을 담당하게되었는데, 특히 연구자 주도 임상시험에서는 대상자 모집하는 과정에 어려움을 느끼고, 또 연구가 조기종료되는 경우들도 발생하기도 합니다. 이 시스템은 이러한 부분에 도움이 될 수 있고, 이에 국내 연구자 주도 임상시험이 더 활성화시킬 수 있다는 점을 기대하는 방향으로 정책을 제안하였습니다.


Q. 졸업 후 진로, 현업과 어떻게 연결할건지

현재는 EDC 시스템을 사용하여 임상시험 데이터를 관리하고 있지만, 본 석사과정을 통해 더 나아가 임상시험의 통계분석 전반을 이해하는 전문가로 성장하고싶습니다.
이를 통해 통계분석을 수행하고, 데이터 관리 경험과 함께 통계적 관점에서 임상연구 설계 하는데에 기여하고 싶습니다.
