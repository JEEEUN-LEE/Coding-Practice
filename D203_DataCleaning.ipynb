```python
import pandas as pd 

fu_df = pd.read_excel('FU,SFU_20250627.xlsx', sheet_name = 'FU_Off Treatment Follow-up')
sfu_df = pd.read_excel('FU,SFU_20250627.xlsx', sheet_name = 'SFU_Survival Follow-up')
print(fu_df.shape, sfu_df.shape)
```

```python
# 조건에 맞는 행 필터링 및 컬럼 추출
fu_filtered = fu_df[
    (fu_df['FUNO'] == 1) &
    (fu_df['FUDT'].notnull()) &
    (fu_df['SUBJ_STATUS'] != 'Study Off')
]

# 첫 번째부터 30번째 컬럼만 추출 (iloc 사용)
fu_filtered = filtered.iloc[:, :30]

# 결과 확인
fu_filtered.head()
```

```python
# SUBJ_ID 중복 확인 및 중복값 출력
duplicates = fu_filtered[fu_filtered.duplicated('SUBJ_ID', keep=False)]
if not duplicates.empty:
    print("중복되는 SUBJ_ID:")
    print(duplicates['SUBJ_ID'].unique())
else:
    print("중복되는 SUBJ_ID 없음")
```

```python
# sfu_df에서 SFUDY 값이 null이 아니고 SUBJ_STATUS가 'Study Off'가 아닌 행만 필터링
sfu_filtered = sfu_df[
    sfu_df['SFUDY'].notnull() & (sfu_df['SUBJ_STATUS'] != 'Study Off')
]
# 결과 확인
sfu_filtered.head()
```

```python
sfu_filtered.columns
```

```python
# 두 데이터프레임을 각각 시트로 저장
with pd.ExcelWriter('filtered_fu_sfu.xlsx') as writer:
    fu_filtered.to_excel(writer, sheet_name='FU_Filtered', index=False)
    sfu_filtered.to_excel(writer, sheet_name='SFU_Filtered', index=False)
```

```python
# FU, SFU 데이터프레임에서 필요한 컬럼만 추출
fu_list = fu_filtered[['LOCATION_NAME', 'STUDY_OID', 'SUBJ_ID', 'SUBJ_STATUS']]
sfu_list = sfu_filtered[['LOCATION_NAME', 'STUDY_OID', 'SUBJ_ID', 'SUBJ_STATUS']]

# 두 데이터프레임을 합치고 SUBJ_ID 기준으로 중복 제거
FU_SFU_List = pd.concat([fu_list, sfu_list], ignore_index=True).drop_duplicates(subset=['SUBJ_ID'])

# 결과에서 LOCATION_NAME, STUDY_OID, SUBJ_ID, SUBJ_STATUS 컬럼만 남김
FU_SFU_List = FU_SFU_List[['LOCATION_NAME', 'SUBJ_ID', 'SUBJ_STATUS']]

# 결과 확인
FU_SFU_List.head()
```

```python
# FU_SFU_List 컬럼명들 변경
# LOCATION_NAME, SUBJ_ID, SUBJ_STATUS를 기관명, 대상자 ID, 대상자 상태로 변경
FU_SFU_List.rename(columns={
    'LOCATION_NAME': '기관명',
    'SUBJ_ID': '대상자 ID',
    'SUBJ_STATUS': '대상자 상태'
} , inplace=True)

# 결과 확인
FU_SFU_List.head()
```

```python
# FU_SFU_List를 엑셀 파일로 저장
FU_SFU_List.to_excel('FU_SFU_List.xlsx', index=False)
```
