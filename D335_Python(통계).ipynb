# 통계적 검정 및 분포 개요

---

## 왜도 (Skewness)

- **왜도(Skewness)**는 분포의 비대칭 정도를 나타낸다.  
- 두 변수 모두 **양수(+)**이면 → 분포가 **왼쪽으로 치우친 분포 (Left-skewed)**  
- 두 변수 모두 **음수(-)**이면 → 분포가 **오른쪽으로 치우친 분포 (Right-skewed)**

---

## 정규성 검정 (Normality Test)

- **Shapiro-Wilk test**의 `p-value`가 **0.05보다 크면**,  
  해당 데이터는 **정규분포를 따른다고 할 수 있다.**

---

## 통계적 검정의 주요 구분

- **정규성 검정**: 정규분포를 따르는가?  
  → 모수/비모수 검정을 구분하는 기준이 됨  
- **왜도 검정**: 어느 쪽으로 치우쳐 있는가?

---

## 모수검정 vs 비모수검정

| 구분 | 모수검정 (Parametric test) | 비모수검정 (Non-parametric test) |
|------|----------------------------|----------------------------------|
| 전제 | 데이터가 **정규분포**를 따른다 | 정규분포를 따르지 않아도 된다 |
| 사용 값 | **평균**, **표준편차** | **중앙값**, **순위(rank)** |
| 특징 | 모집단의 모수(평균, 표준편차)를 이용 | 극단값에 덜 민감 |
| 적용 사례 | 표본이 충분히 큰 경우 (n ≥ 30) | 표본이 작거나 정규성 가정이 어려운 경우 |

---

## t-분포 (t-distribution)

- 모집단의 크기가 작더라도, **정규분포를 가정**하면 표본을 통해 모수를 추정할 수 있다.  
- 모집단의 표준편차 대신 **표본의 불편표준편차(s)**를 사용한다.  
- 따라서, **t-분포는 표본 크기와 관계없이 사용 가능**하며, 보통 정규분포 대신 활용된다.

---

## t-분포를 이용한 양측검정 예시

- **95% 신뢰도 양측 검정**에서  
  20개 표본을 통해 평균이 **27/25가 95% 신뢰구간 안에 존재**하면,  
  귀무가설(H₀)을 **기각할 수 없다.**  
  → 즉, 허리둘레는 변했다고 할 수 없다.

---

## p-value 해석

- **p-value > 0.05**  
  → 귀무가설을 **기각하지 못함**  
  → 검정통계량이 95% 신뢰구간 **안에 있음**

- **p-value < 0.05**  
  → 귀무가설을 **기각함**  
  → 검정통계량이 95% 신뢰구간 **밖에 있음**

---

즉,  
`p-value > 0.05` → 귀무가설 유지  
`p-value < 0.05` → 귀무가설 기각
