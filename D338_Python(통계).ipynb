# 📊 가설검정과 t-test 정리

---

## 🧩 가설검정의 개념

- **p-value를 구하지 않고**, 검정통계량이 신뢰구간 안/밖에 있는지만 본다면  
  → **귀무가설을 기각할 수 있는지 여부만** 판단할 수 있다.

- **p-value를 구하면**,  
  → 유의수준(예: 0.05)보다 크거나 작은지로 귀무가설 기각 여부를 판단할 수 있고,  
  → 동시에 **검정통계량이 얼마나 극단적인 값인지(발생확률이 얼마나 희귀한지)** 알 수 있다.

예를 들어,  
A의 p-value = 0.03, B의 p-value = 0.02라면  
→ B가 더 희귀한 사건이며, 통계적으로 **더 유의하다(more significant)** 고 해석할 수 있다.

📌 정리  
> p-value가 작다 = 발생확률이 낮다 = 통계적으로 더 유의하다 = 검정통계량이 평균에서 멀다

## 🧠 One Sample t-test

- 정규성을 만족하면 모수검정(t-test)을 사용할 수 있고,  
  만족하지 않으면 비모수검정으로 대체한다.

- **상황 예시**  
  모집단 평균(μ₀) = 30  
  표본 평균(x̄) = 27.25  
  → 이 표본이 평균 30인 모집단에서 온 것인지,  
  혹은 다른 평균을 가진 모집단에서 온 것인지 검정한다.

- **가설 설정**  
  - H₀ (귀무가설): 모집단 평균 = 30 → 허리둘레가 변하지 않았다.  
  - H₁ (대립가설): 모집단 평균 ≠ 30 → 허리둘레가 변했다.

- **검정 절차**  
  1. Q-Q Plot으로 데이터의 정규성 확인  
  2. 정규성 검정(Shapiro-Wilk) 수행 — p-value > 0.05이면 정규분포 가정 가능  
  3. One Sample t-test 수행  
  4. p-value > 0.05 → 귀무가설 기각 불가 → “허리둘레가 변했다고 볼 수 없다.”  
     p-value < 0.05 → 귀무가설 기각 → “허리둘레가 변했다.”

## ⚖️ 독립된 두 집단 비교 (Independent Two-Sample t-test)

- **상황 예시**  
  치통 신약 개발 연구에서  
  A 그룹은 대조군, B 그룹은 치료군이라 할 때,  
  두 그룹의 **평균 통증 점수에 차이가 있는지**를 검정한다.

- **가설 설정**  
  - H₀ (귀무가설): 두 집단의 평균은 같다.  
  - H₁ (대립가설): 두 집단의 평균은 다르다.

- **검정 절차**  
  1. 각 집단에 대해 정규성 검정 (Shapiro-Wilk test)  
  2. 등분산성 검정 (Levene’s test)  
  3. 조건 충족 시 Independent Two-Sample t-test 수행  
  4. p-value < 0.05 → 두 집단 간 유의한 차이가 있다고 해석  
     p-value > 0.05 → 두 집단 간 유의한 차이가 없다고 해석

---

## ✅ 요약

| 구분 | 모집단 수 | 사용 목적 | 가정 |
|------|------------|------------|------|
| One Sample t-test | 1개 | 모집단 평균과 표본 평균 비교 | 정규성 |
| Two Sample t-test | 2개 | 두 모집단 평균 비교 | 정규성 + 등분산성 |

---

📍정리
p-value는 단순히 “기각/비기각”의 기준을 넘어서  
**검정통계량이 얼마나 극단적인지, 즉 관측된 결과가 얼마나 희귀한지**를 수치적으로 보여준다.  
따라서 p-value가 작을수록 → 발생확률이 낮고 → 통계적으로 더 유의하며 → 귀무가설로부터 멀리 떨어진 결과임을 의미한다.
