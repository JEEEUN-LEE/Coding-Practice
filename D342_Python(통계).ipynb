# 🔹 검정력(Power)과 표본 크기 계산

---

## 1️⃣ 검정력의 개념

- 실제로 표본 평균이 **모평균과 달라졌을 때**,  
  이를 **달라졌다고 올바르게 결론낼 확률**을 **검정력(Power)** 라고 한다.  

- 검정력과 관련된 요인:
  - **1종 오류(α)와 2종 오류(β)** 발생 확률이 작을수록 → 검정력 ↑  
  - **모집단 분산**이 클수록 → 표본 수 ↑ 필요  
  - **두 표본 평균 간 차이**가 클수록 → 검정력 ↑  
  - **표본 크기**가 클수록 → 검정력 ↑

---

## 2️⃣ 예시: 키 크게 하는 신약 개발

- 모집단 평균 키: **168cm**  
- 모집단 표준편차: **5cm**  
- 신약 투여 후 기대 평균: **169cm**  
- 유의수준: 5% (α = 0.05)  
- 검정력: 80% (1 - β = 0.8)

> 단일 표본 평균 검정(one-sample t-test)을 사용하여 표본 크기 계산

- 효과 크기(effect size, d) 계산:
  
  \[
  d = \frac{\text{두 평균 차이}}{\text{표준편차}} = \frac{169 - 168}{5} = 0.2
  \]

- 양측검정(two-tailed), α = 0.05, 검정력 80% 기준으로 표본 수 계산  
  → 필요한 표본 수: **196명**

---

## 3️⃣ 요약

- **검정력(power)**: 실제로 차이가 있을 때 올바르게 발견할 확률  
- 검정력 ↑: 표본 크기 ↑, 평균 차이 ↑, 모집단 분산 ↓, 1/2종 오류 ↓  
- 표본 수 산출 시 고려 요소: α, β, 효과 크기(d), 양측/단측 검정 여부
