# [보건 통계]

### 1️⃣ 연구방법

- **질적 연구(Qualitative research)**  
  → 주로 **귀납적 방법** 사용  
  → 개별 사실로부터 일반적 원리를 도출  
  - 예시:  
    여러 호랑이를 관찰해보니 모두 검은 줄무늬가 있음 →  
    **“호랑이는 검은 줄무늬가 있는 동물이다.”**

- **양적 연구(Quantitative research)**  
  → 주로 **연역적 방법** 사용  
  → 일반적 원리로부터 개별 사실을 검정  
  - 예시:  
    “호랑이는 검은 줄무늬가 있는 동물이다.” →  
    특정 호랑이의 줄무늬 존재를 예측

---

### 2️⃣ 통계분석 방법의 선택

> 분석 목적과 변수의 종류에 따라 적절한 통계분석 방법을 선택해야 한다.

| 구분 | 주요 분석 방법 | 설명 |
|------|----------------|------|
| **차이 검정 (Difference test)** | - t-검정<br>- 분산분석(ANOVA) | 집단 간 평균 차이를 비교 |
| **관계 검정 (Relationship test)** | - 교차분석<br>- 상관분석<br>- 회귀분석 | 변수 간의 관련성 또는 예측 관계를 분석 |

---

### 3️⃣ 중심극한정리 (Central Limit Theorem)

- 모집단의 분포가 정규분포가 아니더라도,  
  표본의 크기 **n이 충분히 크면**, 표본평균의 분포는 **정규분포에 가까워진다.**
- 즉, 많은 표본을 추출하면 통계적으로 정규분포를 가정할 수 있게 된다.
- 이 원리는 대부분의 통계분석(예: t-test, ANOVA 등)의 **정규성 가정**을 정당화함.

---

### 4️⃣ 정규성 검정 (Normality test)

데이터가 정규분포를 따르는지 확인하기 위한 검정.

- **Kolmogorov–Smirnov test (K–S test)**  
- **Shapiro–Wilk test (S–W test)**

> ✅ 해석 기준  
> - p-value ≥ 0.05 → 정규분포를 따른다고 볼 수 있음  
> - p-value < 0.05 → 정규분포를 따른다고 보기 어려움

# [통계학 기본]

- 통계학은 데이터를 잘 사용하는 방법을 알아내는 학문이라 할 수 있다.
    - 1) 기술 통계: 데이터를 정리 요약하기 위해 사용되는 통계학
    - 2) 추측 통계: 미지의 데이터를 추측하기 위해 사용되는 통계학 

- 확률적으로 변화하는 값을 확률변수라고 한다. 표본을 확률변수라고 간주한다.
- 모집단에서 표본을 얻는 것을 샘플링이라고 한다. 
    - 단순 랜덤 샘플링
